```python
base='..\\data\\brute\\Online_Retail.csv'
data_frame = pd.read_csv(base)
data_frame.info()
data_frame.describe()
data_frame.isnull().sum()
```

supprimer les lignes qui non pas  des CustomerID (identifiant client)  et afficher l'information 
supprimer les doublons 

```python
data_frame=data_frame.dropna(subset=['CustomerID'])
data_frame=data_frame.drop_duplicates()
data_frame.info()
```

separer les commandes annullées et les  commandes vendues
Quantite inferieur à 0['Quantity'])<0 : commande annulée
Quantite superieur à 0  ['Quantity'])>0:commande livrée 

```python
commande_vendue=data_frame[(data_frame['Quantity'])>0]
commande_annule=data_frame[(data_frame['Quantity'])<0]

commande_vendue.describe()
```
